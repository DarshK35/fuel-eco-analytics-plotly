```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Dataset/data_18_v4.csv') 
```

```python
grouped_data = data.groupby(['drive', 'smartway']).size().unstack()
colors = ['#FC5A2E', '#1F77B4','#8BEDEE']
```

```python
grouped_data.plot(kind='bar', stacked=True, color=colors)
plt.xlabel('Drive Types')
plt.ylabel('Vehicle Counts')
plt.legend(title='SmartWay Certification')
plt.title('Distribution of Vehicles by Drive Type and SmartWay Certification')
plt.show()
```

```python
fig=px.treemap(data,path=['veh_class', 'trans'],values='greenhouse_gas_score')
fig.show()
```

```python
columns = ['veh_class']

# Group the data by the selected columns and count the occurrences
grouped_data = data.groupby(columns).size().reset_index(name='count')

# Create the pie chart using Plotly Express
fig = px.pie(grouped_data, names='veh_class', title='Pie Chart of Vehicle Class', values='count')
fig.show()
```

```python
plt.figure(figsize = (18, 10))
plt.bar(grouped_data['veh_class'], height = grouped_data['count'])
plt.show()
```

```python
columns = ['veh_class', 'fuel']

# Count the occurrences of each combination of values
grouped_data = data.groupby(columns).size().reset_index(name='count')

# Create the bar graph using Plotly Express
fig = px.bar(grouped_data, x='veh_class', y='count', color='fuel', title='Bar Graph of Vehical Class and Fuel')
fig.update_yaxes(range=[0, 300], dtick=50)
fig.update_layout(xaxis_title='Vehical class', yaxis_title='Count', legend_title='Fuel')
fig.show()
```
